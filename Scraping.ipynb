```python
import requests 
import bs4 
from bs4 import BeautifulSoup
from urllib.parse import urljoin
import pandas as pd
import numpy as np
```

```python
url_inicial = "https://autos.mercadolibre.com.mx/#deal_print_id=0da78690-c067-11ec-9084-f5ed70aee17d&c_id=carousel&c_element_order=1&c_campaign=AUTOS-Y-CAMIONETAS&c_uid=0da78690-c067-11ec-9084-f5ed70aee17d"
r = requests.get(url_inicial)
r.status_code
```

```python
s = BeautifulSoup(r.text, "lxml") #Este metodo lxml puede variar, se obtiene todo el html de la pagina
```

```python
lista_article = s.find_all("li", class_= "ui-search-layout__item")
lista_article #box de li
link_carros = [x.find("a").get("href") for x in lista_article]
link_carros #links de la pagina
```

```python
def get_url_items(sopa):
    lista_article = sopa.find_all("li", class_= "ui-search-layout__item")
    link_carros = [x.find("a").get("href") for x in lista_article]
    return link_carros
```

```python
links_items = []
for i in s.find_all('a', href = True):
    
  # Checa todo link que contiene "https://autos.mercadolibre.com.mx/"
  if("https://autos.mercadolibre.com.mx/" in i['href']):
      
    # llama al metodo get para solicitar el siguiente url  
    nextpage = requests.get(i['href'])
      
    # crea una sopa para la siguiente url
    nextsoup = BeautifulSoup(nextpage.text, 'lxml')
    
    links = get_url_items(nextsoup)
    links_items.append(links)
    print(links)
```

```python
# Lista de todos los links de cada automoviles
list_scraper = []
for i in links_items:
    for j in i:
        list_scraper.append(j)
len(list_scraper)
```

```python
# Funcion para pbtener los datos en cada una de las paginas
def datos_de_cada_carro(links_de_carros):
    #Obteto request con url
    r_item = requests.get(links_de_carros)

    #Objeto BeautifulSoup con r_item
    file = BeautifulSoup(r_item.text, "lxml")  

    #Encontrar la tabla en la pagina
    rows = file.find('table').find_all('tr')

    # Precio
    precio = file.find("span", class_= "andes-money-amount__fraction" ).get_text()
    
    tabla = [precio]
    
    for row in rows:
        datos = row.find('td').get_text()
        tabla.append(datos)

    return tabla
```

```python
# Generamos el df

encabezado = ['Precio','Marca', 'Modelo', 'Ao', 'Color', 'Tipo_de_combustible', 'Puertas', 'Transmision', 'Motor','Tipo_de_carroceria', 'Km' ]
df = pd.DataFrame(columns = encabezado)
for i in range(4635, len(list_scraper)):
        dato = datos_de_cada_carro(list_scraper[i])
        
        while len(dato)<11:
            dato.append(np.nan)
        
        df=df.append({'Precio' : dato[0],
                      'Marca' : dato[1] ,
                      'Modelo' : dato[2], 
                      'Ao' : dato[3],
                      'Color': dato[4],
                      'Tipo_de_combustible': dato[5],
                      'Puertas' : dato[6],
                      'Transmision' : dato[7],
                      'Motor' : dato[8],
                      'Tipo_de_carroceria' : dato[9],
                      'Km' : dato[10]} , ignore_index=True)   
    
```
